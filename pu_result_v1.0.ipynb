```python
import unittest
import time
import re
import sys
import os
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import Comment
result = []

import csv
```

```python
def grade_to_value(grade):
    if grade == 'S':
        return 10
    elif grade == 'A':
        return 9
    elif grade == 'B':
        return 8
    elif grade == 'C':
        return 7
    elif grade == 'D':
        return 6
    elif grade == 'E':
        return 5
    elif grade == 'F':
        return 0
    else:
        return 0
```

```python
def get_student_result(register_number):
    driver = webdriver.Firefox()
    driver.get("http://result.pondiuni.edu.in/candidate.asp")

    username = driver.find_element_by_id('txtregno')
    username.send_keys(register_number)
    driver.find_element_by_xpath("//select[@name='cmbdegree']/option[text()='BTHCS']").click()
    driver.find_element_by_xpath("//select[@name='cmbexamno']/option[text()='Sixth']").click()
    res={}
    res_sub = ['RegisterNumber']
    res_grade = [register_number]


    driver.find_element_by_id('button1').click()

    html = driver.page_source
    driver.close()
    soup = BeautifulSoup(html)
    for tag in soup.find_all('td',{'width':"66%",'bgcolor':"White"}):
        res_sub.append(tag.text)
    for tag in soup.find_all('font',{'color':"DarkGreen"}):
        grade_num = grade_to_value(tag.text)
        res_grade.append(grade_num)

    res=dict(zip(res_sub,res_grade))
    if len(res)>1:
        result.append(res)
    return 0;
```

```python
def convert_to_csv():
    keys = result[0].keys()
    with open('marks.csv', 'wb') as output_file:
        dict_writer = csv.DictWriter(output_file, keys)
        dict_writer.writeheader()
        dict_writer.writerows(result)
        return 0;
```

```python
input_str_base = '13TD00'

for i in range(1,3):
    if i < 10:
        input_value=input_str_base+"0"+str(i)
    else:
        input_value=input_str_base+str(i)
    print i
    get_student_result(input_value)
    
convert_to_csv()
```
